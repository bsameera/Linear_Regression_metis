```python
individual_page_data(city):
    driver = webdriver.Chrome(chromedriver)
    driver.get(city)
    
    for i in range(5):
    # scroll
    driver.execute_script(
        "window.scrollTo(0, document.documentElement.scrollHeight);" #Alternatively, document.body.scrollHeight
    )
    # page will not load if you don't wait for it
    time.sleep(1)
    
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    
    property_cards = soup.find('div', class_='message-container').find_next_siblings('div', class_='property-card')
    
    starting_prices = []
    links = []

    for property_card in property_cards:
        links.append('https://www.hostelworld.com'+property_card.find('a').get('href'))

        prices = property_card.find_all('div', class_='price title-5')
        if len(prices) == 1:
            for ele in prices:
                #print(ele)
                starting_prices.append(float(ele.text.replace('US$', '').replace('\n', '').strip()[0:3]))
        elif len(prices) == 2:
            price = []
            for ele in prices:
                #print(ele)
                price.append(float(ele.text.replace('US$', '').replace('\n', '').strip()[0:3]))
            starting_prices.append(min(price))
            
    y = 'overall_rating'
    x = ['name', 'starting_price', 'security_rating', 'location_rating', 'staff_rating', 'atmosphere_rating', 'cleanliness_rating', 'facilities_rating', 'value_for_money_rating', 'breakfastincluded', 'freewifi', 'linenincluded', 'lockers', 'airconditioning', 'commonroom', 'laundryfacilities', 'fridgefreezer', 'microwave', 'hotshowers', 'restaurant', 'bar', 'mealsavailable', 'vendingmachines', 'cafe', 'tea&coffeemakingfacilities', '24hourreception', '24hoursecurity']

    list_individual_features = []
    for link in links:
        url = link
        response = requests.get(url)
        status = response.status_code
        if status == 200:
            page = response.text
            soup_link = BeautifulSoup(page)

        individual_features = {'Overall_Rating': 0}

        hostel_name = soup_link.find('div', class_='title-2').text
        individual_features['name'] = hostel_name


        overall_rating = soup_link.find('div', class_='score orange big')
        if overall_rating != None:
            overall_rating = float(overall_rating.text.strip())
            individual_features['Overall_Rating'] = overall_rating


        rating_list = soup_link.find_all('li', class_='rating-item')

        for rating in rating_list:
            rating_text = rating.find('div', class_='rating-label body-3').text.strip()
            rating_value = float(rating.find('div', class_='rating-score body-3').text.strip())
            rating_text = rating_text+'_Rating' 
            individual_features[rating_text] = rating_value

        facilities = soup_link.find_all('ul', class_='facilities')

        all_facilities = []
        for facility in facilities:
            each_facility = facility.find_all('li')
            for ele in each_facility:
                all_facilities.append(ele.text.strip().strip('t_').replace(' ', '').lower())

        features = x[9:]

        for feature in features:
            if feature in all_facilities:
                individual_features[feature] = 1
            else:
                individual_features[feature] = 0

        list_individual_features.append(individual_features)
        
    driver.close()
    
    df = pd.DataFrame(list_individual_features)
    
    prices_df = pd.DataFrame(starting_prices, columns=['starting_prices'])
    
    df2 = pd.concat([df, prices_df], axis=1)

    df2.to_csv('/Users/bsameera/Metis_DS/2_Linear_Regression/Project/hostels_europe.csv', mode='a', header=False)
```

```python
for city in city_links:
    individual_page_data(city)
```
